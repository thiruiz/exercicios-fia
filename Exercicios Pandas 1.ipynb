### 1. Ler e exibir um arquivo CSV usando pandas

```
df = spark.read.csv("/FileStore/tables/retail.csv", header=True, sep=',')
df = df.toPandas()

display(df)
```

### 2. Selecionar colunas especficas de um DataFrame usando pandas

```
df2 = df.filter(['stock_code', 'unit_price'])

display(df2)
```

### 3. Filtrar linhas com base em uma condio usando pandas

```
df3 = df
df3['unit_price'] = df['unit_price'].apply(lambda x: float(x.replace(',', '.')))

df3 = df3.query('unit_price > 10')

display(df3)
```

### 4. Agrupar dados com base em uma coluna usando pandas

```
df4 = df.groupby('region').agg(n=('unit_price', 'count')).reset_index()
display(df4)
```

### 5. Usar a funo max em alguma coluna do DataFrame

```
df['unit_price'].max()
```
